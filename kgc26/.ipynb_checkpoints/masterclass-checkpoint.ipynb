## 2. Data engineering with Polars

The OTTR templates we're about to load expect specific column names like
`?author_iri` and `?name`. Source files use whatever names the original
team picked (`full_name`, `author_id`, `cust_ref`…). `data_engineering.py`
renames source columns and builds full IRIs using **typed segments**
(`.../author/...`, `.../book/...`, etc.) so two different things never
share an identifier.


Harmonising authors! Similar things are happening with books, series and adaptions.

````python
authors = pl.read_parquet("data/authors.parquet")

authors = authors.rename({
    "full_name": "name",     # full_name → name (OTTR template expects "name")
})

# Create IRIs for each author and for their nationality (a country).
authors = authors.with_columns(
    (ns + "author/" + pl.col("author_id")).alias("author_iri"),
    (ns + "country/" + pl.col("nationality")).alias("nationality"),
)

authors = authors.select([
    "name",
    "author_iri", # updated column
    "birth_year",
    "death_year",
    "nationality", # updated column
    "era",
])
````

Books after harmonisation:

### The headline DE moment: a self-join for `:nextInSeries`

To support a Datalog rule that walks the series chain (coming up in section 7),
we need to materialise the "next book in this series" relationship.
That's a classic self-join: rank books within each series, then pair each
book with rank N to the book with rank N+1.

**A perfect example of the T in ETL.**

````python
df = data.books().sort(["series_iri", "position"])

# Add a dense rank within each series (1, 2, 3, …) this gives us
# a clean sequential index regardless of gaps in the source positions.
df = df.with_columns(
    pl.col("position")
      .rank("dense")
      .over("series_iri")
      .cast(pl.Int64)
      .alias("rank"),
)

# The "left" side keeps the current book; the "right" side is the same
# DataFrame, renamed so columns don't collide.
right = df.select([
    pl.col("book_iri").alias("next_book_iri"),
    pl.col("series_iri"),
    (pl.col("rank") - 1).alias("rank"),  # join rank N -> rank N+1
])

pairs = (
    df.select(["book_iri", "series_iri", "rank"])
      .join(right, on=["series_iri", "rank"], how="inner")
      .select(["book_iri", "next_book_iri"])
)

pairs
````

# Time to load data and RDFize it

## 3. OTTR templates

A template describes the shape of triples produced from one DataFrame row. It has a head (template name + typed parameters) and a body (a list of `ottr:Triple(s, p, o)` patterns). Five templates here, one per
DataFrame shape we built above. 

**It's important that the OTTR variable names corresponds with the DataFrame headers!**

To learn more about OTTR, visit the [OTTR spec](https://ottr.xyz/) or the author's [OTTR Masterclass](https://github.com/veleda/ottr-masterclass). 


````python
print(open("tpl/tpl.stottr").read())
````

## 4. The first mapping

Create a `Model`, register the templates, and call `m.map(template_name, df)`. That's the whole pattern!

🔥 **PS: If you don't know how to write OTTR, or want to kick-start your template. Simply do a `m.map_default(df, "subject_iri")`, and maplib will generate a template (as string) and do the mapping for you!**


````python
m = Model()
m.add_template(open("tpl/tpl.stottr").read())

m.map(data.ns_tpl + "Author", data.authors())
print_count("mapping Author", m)
````

Map the rest:

````python
m.map(data.ns_tpl + "Series",           data.series())
m.map(data.ns_tpl + "Book",             data.books())
m.map(data.ns_tpl + "NextInSeriesLink", data.next_in_series())
m.map(data.ns_tpl + "Adaptation",       data.adaptations())

print_count("mapping all", m)
````

Quick sanity check — the graph should already answer questions about
who wrote what:


````python
m.query("""
    PREFIX : <http://data.treehouse.example/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    SELECT ?author_name ?title ?year
    WHERE {
        ?a a :Author ; 
            rdfs:label ?author_name ; 
            :wrote ?b .
        ?b rdfs:label ?title ; 
            :publicationYear ?year .
    }
    ORDER BY ?author_name ?year
""")
````

SKOS gives us controlled vocabularies for genre, era, and adaptation type. DCAT describes the dataset itself — useful when the graph is later published or catalogued.


````python
m.read("ttl/ontology.ttl")
print_count("merge with ontology", m)

m.read("ttl/vocab.ttl")
print_count("merge with SKOS vocabularies", m)

m.read("ttl/catalog.ttl")
print_count("merge with DCAT catalog", m)
````

## 6. Linking string literals to SKOS concepts

Right now a book has `:genre "fantasy"` as a literal string. We'd rather have `:hasGenre :Fantasy`, where `:Fantasy` is a proper SKOS concept that participates in the genre hierarchy. Three SPARQL UPDATEs do the swap: genre on books, era on authors, type on adaptations.


````python
print(open("queries/skos updates/link_genres_to_skos.rq").read())
````

````python
for qf in [
    "queries/skos updates/link_genres_to_skos.rq",
    "queries/skos updates/link_eras_to_skos.rq",
    "queries/skos updates/link_adaptation_types_to_skos.rq",
]:
    m.update(open(qf).read())

print_count("after SKOS linking", m)
````

A book now points at a `skos:Concept` instead of a string. Notice we can
also follow `skos:broader` to climb the genre hierarchy:


````python
m.query("""
    PREFIX :    <http://data.treehouse.example/>
    PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    SELECT ?title ?genre_label ?broader_label
    WHERE {
        ?book a :Book ; rdfs:label ?title ; :hasGenre ?concept .
        ?concept skos:prefLabel ?genre_label .
        OPTIONAL { ?concept skos:broader ?broader . ?broader skos:prefLabel ?broader_label . }
    }
    ORDER BY ?title
""")
````

## 7. Inference: Datalog and/or SPARQL CONSTRUCT

`m.infer()` supports Datalog or SPARQL CONSTRUCT rules for inferencing on data. What differs inference with SPARQL CONSTRUCT from regular SPARQL CONSTRUCTS is the fixed point; deriving new triples until nothing new appears. Recursion comes for free! Same goes for Datalog. Here you have two syntexes, use the one you're the most comfortable with. 


````python
print(open("ttl/rule.dlog").read())
````

Before we run inference, ask the graph a question that *requires* a derived predicate. 

`:writtenBy` is the inverse of `:wrote` — only `:wrote` is in the data, so this should currently return zero rows.


````python
m.query("""
    PREFIX : <http://data.treehouse.example/>
    SELECT (COUNT(*) AS ?writtenBy_count) WHERE { ?b :writtenBy ?a }
""")
````

Run the rules. After this, every implication has been materialised:
inverses, the recursive series propagation, the bridge predicates from
adaptations to authors, and the class-membership tags
(`:AdaptedWork`, `:GoldenAgeBook`, `:ModernEraBook`).

> `m.infer()` is part of maplib enterprise. It's free for academics and
> personal exploration, reach out to any of the Data Treehouse team members.


````python
m.infer(open("ttl/rule.dlog").read())
print_count("inference", m)
````

Same `:writtenBy` query, post-inference, now there are 17 of them:

````python
m.query("""
    PREFIX : <http://data.treehouse.example/>
    SELECT (COUNT(*) AS ?writtenBy_count) WHERE { ?b :writtenBy ?a }
""")
````

**Adapted works.** Every book with at least one adaptation gets tagged
as `:AdaptedWork` by the rule
`[?book, rdf:type, :AdaptedWork] :- [?adaptation, :basedOn, ?book]`.


````python
m.query(open("queries/adapted_books.rq").read())
````

## 10. Write the whole graph to Turtle

Prefix namespaces in a dictionary and pass it on to `m.write`. Supported formats: `"ntriples"`, `"turtle"`, `"rdf/xml"`.

````python
prefixes = {
    "":           "http://data.treehouse.example/",
    "author":     "http://data.treehouse.example/author/",
    "book":       "http://data.treehouse.example/book/",
    "series":     "http://data.treehouse.example/series/",
    "adaptation": "http://data.treehouse.example/adaptation/",
    "country":    "http://data.treehouse.example/country/",
}
m.write("ttl/out.ttl", format="turtle", prefixes=prefixes)

with open("ttl/out.ttl") as f:
    print("".join(f.readlines()[:40]))
print("...")
````
